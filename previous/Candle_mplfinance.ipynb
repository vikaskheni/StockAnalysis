importing library

```python
import mplfinance as mpf
import matplotlib.pyplot as plt
import pandas as pd

import warnings
warnings.filterwarnings('ignore')
```

Using data from CSV file

```python
data = pd.read_csv( '../TCS_stock_data.csv')
print(data)
data['Date'] = pd.to_datetime(data['Date'], format='%d-%m-%Y')
data.set_index('Date', inplace=True)
```

performing slice operation to year 2020

```python
data_2020 = data['2020-01-01':'2020-12-31']
mpf.plot(data_2020)
```

Line Graph

```python
mpf.plot(data_2020, type='line')
```

Candlestick Chart

```python
 data_2023 = data['2023-05-01':'2023-07-01']
 mpf.plot(data_2023, type='candle',title='Stock Open Price in May - July', style='yahoo',figsize=(12, 6))
```

Adding Trading Volume

```python
mpf.plot(data_2023, type='candle', title='Stock Open Price & Volume Data ',style='yahoo',tight_layout=True, volume=True,figratio=(16,9))
#just adding one volume
```

Moving average SMA 5 and SMA 20

```python
mpf.plot(data['2020-07':'2020-12'], type='candle', volume=True,mav=(20,5), title = 'TCS - JAN - DEC(2020)',tight_layout=True, figratio=(16,9))
```
